# Automated Web Scrapping method to collect reviews from all pages of any desired product listed on Amazon Website. 

```python
# Importing necessary libraries for web scrapping 
import requests
from bs4 import BeautifulSoup as bs
from random import randint
from time import sleep
from itertools import repeat

# Creating list to collect the respective outcomes (Names, Reviews, Ratings, Title of Review) from the website.
name= []
cust_name = []
titles = []
review_title = []
ratings = []
rate = []
reviews =[]
review_content = []
modellist = []
mobilemodel =[]
URLS = ['https://www.amazon.in/Test-Exclusive_2020_1151-Multi-3GB-Storage/product-reviews/B089MT35TK/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_revies&pageNumber=']
for URL in URLS: 
    for page in range(2,50):
        pages = requests.get(URL + str(page))
        print(URL)
        print(URL + str(page))
        print(pages)
        soup = bs(pages.content,'html.parser')
        names = soup.find_all('span',class_='a-profile-name')
        name.append(names)
        for i in range(0,len(names)):
            cust_name.append(names[i].get_text())
        title = soup.find_all('a',class_='review-title-content')
        titles.append(title)
        for i in range(0,len(title)):
            review_title.append(title[i].get_text())
        review_title[:] = [titles1.lstrip('\n') for titles1 in review_title]
        review_title[:] = [titles1.rstrip('\n') for titles1 in review_title]
        rating = soup.find_all('i',class_='review-rating')
        ratings.append(rating)
        for i in range(0,len(rating)):
            rate.append(rating[i].get_text()) 
            review = soup.find_all("span",{"data-hook":"review-body"})
            #print("***************************************************")
            #print(review)
        reviews.append(review)
        for i in range(0,len(review)):
            review_content.append(review[i].get_text())
        review_content[:] = [reviews.lstrip('\n') for reviews in review_content]
        review_content[:] = [reviews.rstrip('\n') for reviews in review_content]

#print("Customer names are ",cust_name)
print(len(cust_name))
#print("Review Title is ",review_title)
print(len(review_title))
#print("Number Ratings are ",rate)
print(len(rate))
#print("Actual reviews are ",review_content)
print(len(review_content))
##trying this for getting the model names
#for URL in URLS:
 #   pages = requests.get(URL)
 #   print(URL)
  #  print(pages)
  #  soup = bs(pages.content,'html.parser')
  #  mobile_brands = soup.find("a",{"data-hook":"product-link"})
  #  mobilemodel.append(mobile_brands.get_text())
   # print(mobilemodel)
```

```python
#Importing libraries required to create the dataframe. 
import pandas as pd
df1 = pd.DataFrame()
df1['Customer Name']=cust_name
df2 = pd.DataFrame()
df2['Review title']=review_title
df3 = pd.DataFrame()
df3['Ratings']=rate
df4 = pd.DataFrame()
df4['Reviews']=review_content
```

```python
frames = [df1, df2, df3, df4]

result = pd.concat(frames,axis=1)
```

```python
# Create a new data frame "reviews" by dropping null values to perform exploratory data analysis 
Reviews = result
# Dropping null values
Reviews.dropna(inplace=True)
```

```python
score_1 = Reviews[Reviews['Ratings'] == '1.0 out of 5 stars']
score_2 = Reviews[Reviews['Ratings'] == '2.0 out of 5 stars']
score_3 = Reviews[Reviews['Ratings'] == '3.0 out of 5 stars']
score_4 = Reviews[Reviews['Ratings'] == '4.0 out of 5 stars']
score_5 = Reviews[Reviews['Ratings'] == '5.0 out of 5 stars']
```

```python
reviews_sample = pd.concat([score_1,score_2,score_3,score_4,score_5],axis=0)
reviews_sample.reset_index(drop=True,inplace=True)
```

```python
#Wordcloud function's input needs to be a single string of text.
#Concatenating all Summaries into a single string.
#Similarly you can build for Text column
reviews_str = reviews_sample.Reviews.str.cat()
wordcloud = WordCloud(background_color='white').generate(reviews_str)
plt.figure(figsize=(12,12))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Now let's split the data into Negative (Score is 1 or 2) and Positive (4 or 5) Reviews.
negative_reviews = reviews_sample[reviews_sample['Ratings'].isin(['1.0 out of 5 stars','2.0 out of 5 stars']) ]
positive_reviews = reviews_sample[reviews_sample['Ratings'].isin(['4.0 out of 5 stars','5.0 out of 5 stars']) ]
# Transform to single string
negative_reviews_str = negative_reviews.Reviews.str.cat()
positive_reviews_str = positive_reviews.Reviews.str.cat()
```

```python
wordcloud_negative = WordCloud(background_color='white').generate(negative_reviews_str)
wordcloud_positive = WordCloud(background_color='black').generate(positive_reviews_str)
# Plot
fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(211)
ax1.imshow(wordcloud_negative,interpolation='bilinear')
ax1.axis("off")
ax1.set_title('Reviews with Negative Ratings',fontsize=20)
```

```python
fig = plt.figure(figsize=(15,15))
ax2 = fig.add_subplot(212)
ax2.imshow(wordcloud_positive,interpolation='bilinear')
ax2.axis("off")
ax2.set_title('Reviews with Positive Ratings',fontsize=20)
plt.show()
```

```python
import seaborn as sns
import re
import os
import sys
import ast
plt.style.use('fivethirtyeight')
# Function for getting the sentiment
cp = sns.color_palette()
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Generating sentiment for all the reviews present in the dataset
emptyline=[]
for row in result['Reviews']:
    
    vs=analyzer.polarity_scores(row)
    emptyline.append(vs)
# Creating new dataframe with sentiments
data_sentiments=pd.DataFrame(emptyline)
data_sentiments.head()
```

```python
# Merging the sentiments back to reviews dataframe
data_merged = pd.concat([result.reset_index(drop=True), data_sentiments], axis=1)
data_merged.head(6)
```

```python
# Convert ratings into positive and negetive sentiments using some threshold
data_merged['Sentiment'] = np.where(data_merged['compound'] >= 0 , 'Positive','Negative')
data_merged.head(8)
```

```python
Result=data_merged['Sentiment'].value_counts()
Result.plot(kind='bar', rot=0, color=['plum','cyan']);
```
